```python
import polars as pl
import os
```

```python
mod_codes = [
    'a',
    'm',
    '17802',
    '17596',
    '19227',
    '19228',
    '19229',
    '69426'    
]

columns = [
    'Chromosome',
    'Start',
    'End',
    'Call',
    'Score',
    'Strand',
    'Start_1',
    'End_1',
    'RBG',
    'N_valid_cov',
    'Mod_Percent',
    'N_mod',
    'N_canon',
    'N_other_mod',
    'N_delete',
    'N_fail',
    'N_diff',
    'N_no_call',
    'IVT_Adjusted_Mod_Percent'
]
```

```python
def filter_and_output_parquet_files(lf, mod_code, outdir):
    df = lf.filter(
        (pl.col('Call') == mod_code) &
        (pl.col('Score') >= 20) & 
        (pl.col('IVT_Adjusted_Mod_Percent') >= 20)
    ).collect()

    if outdir is not None:
        outpath = os.path.join(outdir, f"Dorado_filtered_20_20_{mod_code}.parquet")
        df.write_parquet(outpath)
```

```python
pileup_path = "../../Exemplar_Data/modkit/08_07_24_R9RNA_GM12878_mRNA_RT_sup_8mods_polyA_sorted_filtered.chr12-112000000-114000000_pileup_fp_adjusted.tsv"
outdir = "../../Exemplar_Data/parquet_files/"
lazyframe = pl.scan_csv(pileup_path, separator = "\t", new_columns = columns)

for code in mod_codes:
    filter_and_output_parquet_files(lazyframe, code, outdir)
```
